### 기본 그래프

    막대 그래프(bar), 선 그래프(line), 산포도(scatter), 상자(box), 히스토그램
    
    1) 질적 데이터
        - 요약 방법 : 도표
        - 정리 방법 : 도수 분포표. 분할표
        - 그래프 : 막대 그래프, 원 그래프, ...
        
    2) 양적 데이터
        - 요약 방법 : 수치
        - 정리 방법 : 산술평균, 중앙값,...
        - 그래프 : 히스토그램, Boxplot, 시계열 그래프, 산포도 (산점도)
        
        

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as pit
import seaborn as sns

import matplotlib.font_manager as fm
font_name= fm.FontProperties(fname="C:\\Windows\\Fonts\\malgun.ttf").get_name()
pit.rc('font', family=font_name)


mpl.rcParams["axes.unicode_minus"]=False
```

### 선그래프

```python
s = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0.100,10))
#s.plot()
s.plot.line()
```

```python
df1 = pd.DataFrame(np.random.randn(10,4), columns=["가","나","다","라"], index = np.arange(0,100,10))
df1
df1.plot()
```

```python
df1["가"].plot()
```

```python
s2 = pd.Series(np.random.randn(16), index = list('abcdefghijklmnop'))
# s2.plot(kind = 'bar')
s2.plot.bar()
```

```python
s2.plot(kind="barh")
```

```python
df2 = pd.DataFrame(np.random.randn(6,4), index = ['one', 'two', 'three', 'four', 'five', 'six'], 
                   columns = ['A', 'B', 'C', 'D'])
df2.plot(kind = 'bar')
```

```python
df2 = pd.DataFrame(np.random.randn(6,4),
index=['one','two','three','four','five','six'],
columns=pd.Index(['A','B','C','D'],name='GNP'))
df2.plot.bar()
```
